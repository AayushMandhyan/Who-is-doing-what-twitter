```python
data = open('../data/2016-03-08.json')
d = data.readlines()
tweets = [json.loads(t) for t in d]
```

```python
hash_cluster = json.loads(open('../result/hash_cluster.json').read())
hash_cluster
```

```python
for tweet in tweets:
    for hashtags in tweet['entities']['hashtags']:
        hash = hashtags['text']
        for key, value in hash_cluster.items():
            if hash in value:
                tweet_cluster[key].append(tweet)
```
